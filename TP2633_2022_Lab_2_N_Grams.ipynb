### **Frequency Distribution**
Before we get into Ngrams, it is good to know some functions to display the results.
One of the useful functions is Frequency Distribution


Frequency distribution tells us the frequency of each vocabulary item in the text (or any observable event).
It is a "distribution" because it tells us how the total number of word tokens in the text are distributed across the vocabulary items.


Useful functions to be used where applicable:

Let's try a simple program that display the result of the frequency of each unique word (including symbols) in tabular format using **FreqDist**

```
import nltk
nltk.download('punkt')
from nltk.probability import FreqDist

text1 = "I am Sam. Sam I am. "
tokens = nltk.word_tokenize(text1)
fdist1 = FreqDist(tokens) #use of frequency distribution
print (fdist1.tabulate())
```

### **Use of FreqDist with bigrams**

```
import nltk
from nltk import bigrams


string = "I really like Python I really like Python yes"
tokens = nltk.word_tokenize(string)

#create your bigrams
mybigram = nltk.bigrams(tokens)

#compute frequency distribution for all the bigrams in the text
fdist = nltk.FreqDist(mybigram)
for k,v in fdist.items():
    print (k,v)
```

# **Calculating bigram probabilities**

We can calculate bigram probabilities using some predefined functions in Python.   
Try the following code

```
import nltk
corpus = """<s> I am Sam </s>
<s> Sam I am </s>
<s> I do not like green eggs and ham </s>"""

words = corpus.split()
cfreq_sam = nltk.ConditionalFreqDist(nltk.bigrams(words))
print(cfreq_sam)
cprob_sam = nltk.ConditionalProbDist(cfreq_sam, nltk.MLEProbDist)
print(cprob_sam)
print (cprob_sam["<s>"].prob("Sam")) #probability of P(Sam|<s>)

```

**Question 2**: What is the probability of P(am|Sam)? 

**Answer:** 

P(am|Sam) = count(Sam, am)/ count(Sam)

```
import nltk
corpus = """<s> I am Sam </s>
<s> Sam I am </s>
<s> I do not like green eggs and ham </s>"""

words = corpus.split()
cfreq_sam = nltk.ConditionalFreqDist(nltk.bigrams(words))
cprob_sam = nltk.ConditionalProbDist(cfreq_sam, nltk.MLEProbDist)
print (cprob_sam["Sam"].prob("am")) #probability of P(am|Sam)
```

Amend the program so that unigrams counts are correct. You need to do the following amendments:

*   Make Unigrams[word] = Unigrams[word]+1 if we have seen this before

Hint: Use  **if...else** statement

Print the output of the word and its count in a presentable format

Find the probability of **P(eat)** and **P(box)**


```
#write your program here
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

myfile = open('/content/drive/My Drive/rhyme.txt')
mylines = myfile.readlines()

count = 0
Unigrams = {} #creating a dictionary called Unigrams
for line in mylines:
  #Open and read file
    line = line.rstrip() #removes whitespace at the end
   
#tokenize the text
    tokens = word_tokenize(line)
    
#loop over unigrams
    for word in tokens:
      count += 1
      if word not in Unigrams:
        Unigrams[word] = 1
      else:
        Unigrams[word] += 1 
print(Unigrams)

# Probability of eat
eat = Unigrams["eat"]/count 
print("Probability of eat = ", eat)

# Probability of box
box = Unigrams["box"]/count 
print("Probability of box = ", box)

myfile.close()
 
```
