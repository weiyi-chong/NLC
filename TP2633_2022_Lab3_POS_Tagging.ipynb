# Chunking

Chunking basically means grouping words together into meaningful chunks. It is also known as shallow parsing.

In POS tagging, we might like to chunk words into noun phrases or verb phrases.

In order to chunk, we combine part of speech tags with **regular expression**.


Do you still remember regex?

Refer to your previous lecture notes on regex. 
Basically, we want to identify the POS tags patterns and utilise the following :

Screen Shot 2020-11-07 at 1.37.22 PM.png

In this exercise, we are using the raw string notation (r‘text’) rather than re command for regex.

Raw string notation (r’text’) keeps regular expressions meaningful and confusion-free.
Without it, every backslash ('\\') in a regular expression would have to be prefixed with another one to escape 

Let's have a look again at how chunking can be used to identify the noun phrase in the previous program:

```
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer

train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")

custome_sent_tokenizer = PunktSentenceTokenizer(train_text)

tokenized = custome_sent_tokenizer.tokenize(sample_text)

def process_content():
    try:
        for i in tokenized[:5]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            #the following is  a pattern to find the occurrences of all nouns using regex
            chunkGram = r''' NOUN_FOUND: {<NN.?>*}''' 
            chunkParser =nltk.RegexpParser(chunkGram)
            chunked =chunkParser.parse(tagged)
            print (chunked)
            
    except Exception as e:
            print(str(e))
            
process_content()
```

What did you observe?
**When nouns are found, it will print NOUN_FOUND and the continuous nouns will be grouped together**

Try using other sentences

Try the following patterns:


<RB.?>* = "0 or more of any tense of adverb” 

<VB.?>* = "0 or more of any tense of verb” 

<NNP>+ = "One or more proper nouns”

<NN>? = "zero or one singular noun.”

Try to combine all the patterns like this:

**chunkGram** = r‘‘‘Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?}‘‘‘
